```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price


class Seller:
    def __init__(self, name, shop_name):
        self.name = name
        self.shop_name = shop_name
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def show_products(self):
        print("\n------ PRODUCT LIST ------")
        for i, p in enumerate(self.products, start=1):
            print(f"{i}. {p.name} - ₹{p.price}")


class Customer:
    def __init__(self, name, phone, address):
        self.name = name
        self.phone = phone
        self.address = address
        self.cart = []

    def add_to_cart(self, product, qty):
        self.cart.append((product, qty))
        print(f"{product.name} x{qty} added to cart ✅")

    def generate_bill(self, seller):
        print("\n=========== BILL ===========")
        print(f"Shop Name : {seller.shop_name}")
        print(f"Seller    : {seller.name}")
        print("----------------------------")
        print(f"Customer  : {self.name}")
        print(f"Phone     : {self.phone}")
        print(f"Address   : {self.address}")
        print("----------------------------")

        total = 0
        print("\nItems Purchased:")
        for product, qty in self.cart:
            price = product.price * qty
            total += price
            print(f"{product.name} x{qty} = ₹{price}")

        print("----------------------------")
        gst = total * 0.05
        final_total = total + gst

        print(f"Subtotal : ₹{total}")
        print(f"GST (5%) : ₹{gst}")
        print(f"Total    : ₹{final_total}")
        print("=========== THANK YOU ===========")


# Main Program
def main():
    # Seller details
    seller = Seller("KUSHAL PRAJAPATI", "JENA MART")

    # Add 10+ products
    product_list = [
        Product("Laptop", 50000),
        Product("Mobile", 20000),
        Product("Headphones", 1500),
        Product("Keyboard", 800),
        Product("Mouse", 500),
        Product("Monitor", 12000),
        Product("Printer", 7000),
        Product("Tablet", 15000),
        Product("Camera", 30000),
        Product("Speaker", 2500)
    ]

    for p in product_list:
        seller.add_product(p)

    # Customer input
    print("Enter Customer Details:")
    name = input("Name: ")
    phone = input("Phone: ")
    address = input("Address: ")

    customer = Customer(name, phone, address)

    # Shopping
    while True:
        seller.show_products()
        choice = int(input("Select product number (0 to finish): "))

        if choice == 0:
            break

        if 1 <= choice <= len(seller.products):
            qty = int(input("Enter quantity: "))
            customer.add_to_cart(seller.products[choice - 1], qty)
        else:
            print("Invalid choice ❌")

    # Generate Bill
    customer.generate_bill(seller)


# Run program
main()
```
